```
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import Row
```

```
def insertQuery(dataFrame, database, collection):
    dataFrame.write.format('com.mongodb.spark.sql.DefaultSource').mode("append").option("database", database).option("collection", collection).save()




    
```
